set A is a subset of a set B if all elements of A are also elements of B; B is then a superset of A. It is possible for A and B to be equal; if they are unequal, then A is a proper subset of B

```
#Power Set

def DPpowerset(s):
  powerset = [[]]
  for i in s:
    for sett in powerset:
      powerset = powerset + [sett + [i]]
  return powerset
```

```
DPpowerset(A)
```

```
def powerset(s):
    x = len(s)
    masks = [2**i for i in range(x)]
    for i in range(2**x):
        yield [ss for mask, ss in zip(masks, s) if i & mask]

print(list(powerset(A)))
```
